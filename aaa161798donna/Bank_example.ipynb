```python
# for classification task
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bank = pd.read_csv("bank-full.csv",delimiter=';')
```

```python
bank.head()
```

```python
bank.info()
```

```python
bank.shape
```

```python
bank['y']=bank['y'].map({'yes':1,'no':0})
```

```python
import statsmodels.formula.api as smf
```

```python
string_cols = ' + '.join(bank.columns[:-1])
results = smf.logit('y ~ {}'.format(string_cols), data=bank).fit()
results.summary()
```

```python
# using sk learn

df_dummy=pd.get_dummies(bank,drop_first=True) #remove the first variable that being created. Total variables is n.
df_dummy.head()
# encounter categorical variables, manual convert the variables 
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
lr = LogisticRegression()
X = np.column_stack((np.ones(len(df_dummy)), df_dummy.drop(columns=['y']))) # create the confusion matrix, and drop the response variable
```

```python
X
```

```python
lr.fit(X,df_dummy['y'])
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(df_dummy['y'],lr.predict(X))
```
